# Fixing file encodings

Some of the files in the `data` directory are encoded in the `ANSI` format. This is not supported in unix, so they need to be converted.

```python
import pandas as pd
```

```python
sdg_data = pd.read_csv("../data/sdg_data.csv", encoding = 'ansi', 
                       low_memory = False)

selected_indicators = pd.read_csv("../data/selected_indicators.csv", 
                                  encoding = "ansi")

country_data = pd.read_csv("../data/country_metadata.csv", encoding = "ansi")
```

```python
sdg_data.to_csv('../data/sdg_data.csv', encoding='utf-8', index=None)
selected_indicators.to_csv('../data/selected_indicators.csv', encoding='utf-8', index=None)
```

```python
country_data.to_csv('../data/country_metadata.csv', encoding='utf-8', index=None)
```
