```python
country_data = pd.read_csv("country_metadata.csv", encoding="ansi")
```

```python
def parse_population(n):
    num_part = float(str(n).split(" ")[0])
    if "billion" in n:
        return num_part * 1e9
    elif "million" in n:
        return num_part * 1e6
    elif "thousand" in n:
        return num_part * 1e3
    elif "hundred" in n:
        return num_part * 1e2
    else:
        return num_part
```

```python
country_data.loc[~country_data.population.isnull(), 'population'] = country_data[~country_data.population.isnull()].population.apply(parse_population)
```
